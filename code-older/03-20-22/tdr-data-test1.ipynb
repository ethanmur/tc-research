```python
# import data using xarray
inbound_data = xr.open_dataset('210827H1_2118_vert_inbound.nc')
outbound_data = xr.open_dataset('210827H1_2118_vert_outbound.nc')

# view the data
# outbound_data
# print( inbound_data)
```

```python
# Plotting method #1: using the built in xarray functions
# this method seems quicker but not as customizable or intuitive as using matplotlib

# inbound_data.REFLECTIVITY.transpose().plot()
```

```python
# find the highest altitude and use that as the max y height!

alt_in = inbound_data.variables['altitude']
max_alt_in = alt_in.max().values

alt_out = outbound_data.variables['altitude']
max_alt_out = alt_out.max().values

print( inbound_data.variables['radius'].max().values)
```

```python
# accessing data via variable assignment
# refl_in = inbound_data.variables['REFLECTIVITY']
# refl_out = outbound_data.variables['REFLECTIVITY']
# r_out = outbound_data.variables['radius']

# Plotting method #2: using matplotlib, and field access and isel filters from xarray
fig = plt.figure( figsize=(16, 12))
plt.subplot(211)
color_map = plt.cm.get_cmap( "RdYlBu").reversed()

# make a red line through the plot to show flight height
xvals = np.linspace( 0, 140, 100)
yvals = max_alt_in * np.ones(( 100))
plt.plot( xvals, yvals, color='red', linewidth=3)
                             
plt.pcolormesh( inbound_data.radius, inbound_data.height, inbound_data.REFLECTIVITY.isel(time=0).isel(heading=0).transpose(), cmap = color_map )
plt.title( '2D Cross Section of Reflectivity in Hurricane Ida (8/27/21)')
plt.colorbar( label="Reflectivity (dBZ)")
plt.ylabel( 'Height from Surface (km)')
plt.xlim( [0, 140])
plt.grid( 'on')

plt.subplot(212)
color_map = plt.cm.get_cmap( "RdYlBu").reversed()

plt.pcolormesh( inbound_data.radius, inbound_data.height, inbound_data.REFLECTIVITY.isel(time=0).isel(heading=0).transpose(), cmap = color_map )
plt.title( 'Reflectivity Below 3 km')
plt.colorbar( label="Reflectivity (dBZ)")
plt.xlabel( 'Distance from TC Center (km)')
plt.ylabel( 'Height from Surface (km)')
plt.ylim( [0, max_alt_in])
plt.xlim( [0, 140])
plt.grid( 'on')
```

```python
# the same but for outbound data
# data above flight level is intentionally being ommitted here

fig = plt.figure( figsize=(16,4))
color_map = plt.cm.get_cmap( "RdYlBu").reversed()

plt.pcolormesh( outbound_data.radius, outbound_data.height, outbound_data.REFLECTIVITY.isel(time=0).isel(heading=0).transpose(), cmap = color_map )
plt.title( '2D Cross Section of Reflectivity in Hurricane Ida (8/27/21)')
plt.colorbar( label="Reflectivity (dBZ)")
plt.xlabel( 'Radius (km)')
plt.ylabel( 'Height (km)')
plt.xlim( [-10, 260])
plt.ylim( [0, max_alt_out])

plt.grid( 'on')
```

```python
# check if there is any difference between inbound, inbound rel, and inbound fall data

inbound_rel = xr.open_dataset('210827H1_2118_vert_inbound_rel.nc')
inbound_fall = xr.open_dataset('210827H1_2118_vert_inbound_fall.nc')

refl = inbound_data.REFLECTIVITY.isel(time=0).isel(heading=0).transpose()
refl_rel = inbound_rel.REFLECTIVITY.isel(time=0).isel(heading=0).transpose()
refl_fall = inbound_fall.REFLECTIVITY.isel(time=0).isel(heading=0).transpose()

delta_rel = refl - refl_rel
delta_fall = refl - refl_fall

print( "What is the max difference between inbound reflectivity and 1) inbound rel and 2) inbound fall dBZ values?")
print( delta_rel.max( ).values)
print( delta_fall.max( ).values)
print("Visually:")

fig = plt.figure( figsize=(12,7))
# color_map = plt.cm.get_cmap( "RdYlBu").reversed()

plt.pcolormesh( inbound_data.radius, inbound_data.height, delta_rel) # , cmap = color_map )
plt.title( '2D Cross Section of Reflectivity in Hurricane Ida (8/27/21)')
plt.colorbar( label="Reflectivity (dBZ)")
plt.xlabel( 'Radius (km)')
plt.ylabel( 'Height (km)')
plt.xlim( [0, 140])
plt.grid( 'on')
```
