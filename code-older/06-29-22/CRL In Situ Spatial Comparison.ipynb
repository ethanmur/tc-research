```python
# import packages and plotting scripts

import numpy as np
import os
import matplotlib.pyplot as plt
import xarray as xr
from datetime import datetime
import warnings
import pandas as pd
import cartopy.crs as ccrs
import shapely.geometry as sgeom
import matplotlib.patches as mpatches
from scipy.signal import find_peaks
import cartopy.feature as cfeature
import math


os.chdir(  "/Users/etmu9498/research/code/scripts")
import make_plots
import goes_gifs
```

```python
# view datasets

# paths to data
flight_path = "/Users/etmu9498/research/data/in-situ"
crl_path = "/Users/etmu9498/research/data/CRL_data/2021"
# tdr_list = make_plots.load_tdr(tdr_path)
crl_list = make_plots.load_crl(crl_path)
flight_list = make_plots.load_flight_level( flight_path )
```

```python
# load crl and tdr data
os.chdir( crl_path)
crl_name = crl_list[ 17]
crl_data = xr.open_dataset( crl_name)

# load in situ data
os.chdir( flight_path)
in_situ_data = pd.read_csv( flight_list[ 32], header=None)
in_situ_data


for col_ind in range( len( in_situ_data.columns)):
    # print( col_ind)
    if in_situ_data.iloc[ col_ind][1] == 'TIME':
        in_situ_data.columns = in_situ_data.iloc[ col_ind]
        break

# removing 1203 rows that just repeat the header keys, and 6 rows with empty lat lon
in_situ_data = in_situ_data[ in_situ_data['TIME'] != 'TIME']
in_situ_data = in_situ_data[ ~ pd.isna( in_situ_data['LATref']) ]
# removing 4 columns that are labeled as 'none'
in_situ_data.drop( 'none', inplace=True, axis=1)

# trim out every 10th element for a new, smaller dataset. Easier / faster to work with?
in_situ_data_trim = in_situ_data.iloc[ ::10, :]
# reset indices so they're nice and pretty!
in_situ_data_trim = in_situ_data_trim.reset_index( drop=True)
# in_situ_data_trim

# adding datetime and just time columns to pandas dataframe
in_situ_data_trim['dt'] = pd.to_datetime( in_situ_data_trim['TIME'])
in_situ_data_trim['time'] = [dt_object.time() for dt_object in in_situ_data_trim.dt ]

# convert from pandas to xarray
xr_in_situ =  pd.DataFrame.to_xarray( in_situ_data_trim) 
# remove 'index' from the coordinates tab... not sure about this?
xr_in_situ = xr_in_situ.drop_vars( 'index')

xr_in_situ

crl_data
# **** intersting parameters to look at:
# Note: * means that variable isn't found in .txt file but is found in larger .nc file from noaa website

# ASfmrRainRate.1 = estimated rain rate, maybe see rates along bands?
# SfmrTB = brightness temperature in storm!
# lwc = liquid water content *
# gs = ground speed, m/s
# tas = true air speed
# ias = indicated air speed
# ws = wind speed m/s
# wd = wind direction
# uwz = vertical wind m/s
# ps = corrected static pressure
# pq = corrected dynamic pressure
# dv = d value? not sure what that is...
# mr = mixing ratio
```

```python
# get data out of xarray and put it in a useable format
time = xr_in_situ.time.values
str_time = [ ti.strftime('%H:%M:%S') for ti in time ]
float_time = []
for val in time:
    # account for wraparound times into next day
    if val.hour + val.minute / 60 + val.second / 3600 < 6.0:
        float_time.append( 24.0 + val.hour + val.minute / 60 + val.second / 3600)
    else:
        float_time.append( val.hour + val.minute / 60 + val.second / 3600)

pitch = xr_in_situ.PITCHref.values
pitch = [ float( line) for line in pitch]

roll = xr_in_situ.ROLLref.values
roll = [ float( line) for line in roll]
```

```python
# timestep = .1
# time_int = int( ( crl_data.time[-1] - crl_data.time[0] ) / timestep )

time_int = 10
time_array = np.linspace( crl_data.time[0], crl_data.time[-1], time_int)
print( time_array)
```

```python
for tval in time_array:

    # make figure: setup
    fig = plt.figure(figsize=(16, 10))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    
    # ax.set_extent([-52.5,-48.5, 12.5, 16.5], crs=ccrs.PlateCarree())
    # ax.set_extent([-56.5, -48, 12, 20], crs=ccrs.PlateCarree())
    ax.set_extent([-55.5, -50.5, 14.5, 18.5], crs=ccrs.PlateCarree())

    gl = ax.gridlines( draw_labels=True) #, color="k")
    gl.top_labels = False
    gl.right_labels = False

    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.OCEAN)
    ax.add_feature(cfeature.COASTLINE)

    crllat = crl_data.Lat
    crllon = crl_data.Lon    
    crltrack = sgeom.LineString(zip( crllon, crllat))
    ax.add_geometries([crltrack], ccrs.PlateCarree(),
                      facecolor='none', edgecolor='b' ) # , marker='-')                          

    flightlat = xr_in_situ.LATref
    flightlon = xr_in_situ.LONref    
    flighttrack = sgeom.LineString(zip( flightlon, flightlat))
    ax.add_geometries([flighttrack], ccrs.PlateCarree(),
                      facecolor='none', edgecolor='g')                          


    r = mpatches.Rectangle((0, 0), 1, 1, facecolor="b")
    r2 = mpatches.Rectangle((0, 0), 1, 1, facecolor="g")
    labels = ['crl flight', 'in situ path']
    ax.legend( [r, r2], labels)

    ax.set_title( "Flight Pattern Collocation: CRL vs In Situ")
    # plt.show()
    
    # print( "timestep: " + str( tval))

    # plot star showing plane position: crl
    for ind_t in range( len( crl_data.time)):
        
        # print( round( float( crl_data.time[ ind_t] ), 2))
        if np.isclose( round( float( crl_data.time[ ind_t] ), 3), round( tval, 3), rtol=.00001 ):
            ax.scatter( crl_data.Lon[ ind_t], crl_data.Lat[ ind_t], s=40, c= 'b', marker='o', transform= ccrs.PlateCarree() )
            print('crl time: ' + str( float( crl_data.time[ ind_t] ) ))
            print('crl lon: ' + str( crl_data.Lon[ ind_t].values)) 
            print('crl lat: ' + str( crl_data.Lat[ ind_t].values))
            break
    
    
    # plot smaller circle showing plane position: in situ
    for ind_t in range( len( float_time)):

            # print( round( float( crl_data.time[ ind_t] ), 2))
            if np.isclose( round( float( float_time[ ind_t] ), 3), round( tval, 3 ), rtol=.0001 ):
                ax.scatter( float( xr_in_situ.LONref[ ind_t].values), float( xr_in_situ.LATref[ ind_t].values), s=40, c= 'g', marker='o', transform= ccrs.PlateCarree() )
                print('in situ time: ' + str( float( float_time[ ind_t] ))) 
                print('in situ lon: ' + str( xr_in_situ.LONref[ ind_t].values)) 
                print('in situ lat: ' + str( xr_in_situ.LATref[ ind_t].values) + '\n')
                
                break
```
