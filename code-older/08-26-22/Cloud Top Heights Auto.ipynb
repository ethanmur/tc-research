```python
crl_path = "/Users/etmu9498/research/data/CRL_data/2021"
crl_list = make_plots.load_crl( crl_path)
crl_name = crl_list[ 16]
```

```python
# make simple line plots showing power distributions vs height for a single cross section

cutoff_power = -30
i1= 0
i2= 1600

os.chdir( crl_path)
data = xr.open_dataset( crl_name)
power = data.P_ch1[ i1: i2]

time = data.time[ i1: i2]

H = data.H

# cut off top 8 values to get rid of NaNs from flight level data
power = power[:, 8:]
H = H[ 8:]

# convert to dBz
power = 10 * np.log10( power)

power = power.where( power.values > cutoff_power)
power = np.where( np.isnan( power) , -200, power)

time_ind = 1200 # 1200 # 722 # 755 # 655 # 1147

print( len( H))
print( np.size( power, 1))

# prominence tells the function to look for peaks based on valleys around it.
# this is good at cutting through to actual peaks! see saved stack overflow post.
# wlen tells find_peaks to only choose values within a given range for prominence calculations.
# this gets rid of rain backscatter, as those signals have wide but gradual prominence profiles.
# width makes sure that there are nearby, tall points around the peak.
# this gets rid of aerosol backscatter, as these returns are extremely sharp / narrow.
peaks = find_peaks( power[ time_ind, :], prominence= 5, wlen=100, width=5)
print( peaks[0])
print( peaks[1])

plt.figure( figsize=(8, 8))
plt.plot( power[ time_ind, :], - H )
plt.scatter( power[time_ind, peaks[0]], - H[ peaks[0]], label='No Precip or Aerosol', marker='x', c='r')
plt.title( "Backscattered Power vs Height for TC Sam, 9/26 at " + str( time[ time_ind].values) + " UTC")
plt.ylabel( "Height from Surface (Km)")
plt.xlabel( "Backscattered Power (dBz)")
plt.xlim( [-50, 0])
plt.grid('on')
plt.legend(loc='upper right')
```
