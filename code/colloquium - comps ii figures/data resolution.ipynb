```python
os.chdir("/Users/etmu9498/research/data/crl-all-data-processed/2021/")
crl = xr.open_dataset("P3_20210927H1_processed.nc")

steps = 50
timehr = (crl.time.values[steps] - crl.time.values[0]) / steps
timesec = timehr * 3600
print( "CRL Timestep: {} seconds".format(timesec) )

steps = len( crl.height) - 1
ht = (crl.height.values[steps] - crl.height.values[0]) / steps
print( "CRL Vertical bins: {} m".format(ht) )

crl
```

```python
# processed fl data
os.chdir("/Users/etmu9498/research/data/in-situ-noaa-full/2021/")
fl = xr.open_dataset("20210926H1_sam.nc", decode_times=False) # 20210927H1_sam


plt.plot(fl.Time, fl['GS.d'], c='y', label='ground speed')
plt.plot(fl.Time, fl['IAS.d'], c='c', label='indicated air speed')
plt.plot(fl.Time, fl['TAS.d'], c='g', label='true air speed')

plt.axhline(y=130, c='b')
plt.axvline(x=15500)
plt.axvline(x=28000)

spd = np.nanmean(fl['TAS.d'].values[ 15500:28000])
print( "FL Speed: {} m/s".format(spd) )
```
