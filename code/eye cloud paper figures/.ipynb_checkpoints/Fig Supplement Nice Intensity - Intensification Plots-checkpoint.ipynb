```python
import os
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

os.chdir("/Users/etmu9498/research/code/scripts-winter2023")
import helper_fns_winter2023
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/cloud-top-height-stats")
import eyewall_metadata
```

```python
defined = False

# get 2021 and 2022 intensity and intensificaiton data!
md = eyewall_metadata.all_metadata()

# pull and combine data
intensity2021, intensity2022 = list( md['2021']['intensity'].values()), list( md['2022']['intensity'].values())
intr2021, intr2022 = list( md['2021']['intensification'].values()), list( md['2022']['intensification'].values())
intensity = intensity2021 + intensity2022
ir = intr2021 + intr2022
intensity = np.array(intensity)
ir = np.array(ir)
# print(intensity)
# print(ir)


# 6/5/23 new code: repeat for shear magnitude and direction
sm21, sm22 = list( md['2021']['shear_mag'].values()), list( md['2022']['shear_mag'].values())
sd21, sd22 = list( md['2021']['shear_dir'].values()), list( md['2022']['shear_dir'].values())
shearmag = sm21 + sm22
sheardir = sd21 + sd22
shearmag = np.array(shearmag)
sheardir = np.array(sheardir)

# an optional "side goal": only create distributions for tcs with "defined eyewalls"
# get and combine lists of defined eyewalls
defined2021, defined2022 = list( md['2021']['defined_eyewall'].values()), list( md['2022']['defined_eyewall'].values())
defined_eyewalls = defined2021 + defined2022


if defined:
    intensity = intensity[ np.where(defined_eyewalls)[0]]
    ir = ir[ np.where(defined_eyewalls)[0]]
    
# break down intensity and intensifications into further categories to make different colored bars!
i_td = intensity[intensity < 34.]
i_ts = intensity[np.logical_and( intensity >= 34., intensity < 64.)]
i_wh = intensity[np.logical_and( intensity >= 64., intensity < 96.)]
i_sh = intensity[intensity >= 96.]

# intensification compositing
irlimit = 10
ir_weak = ir[ir < -irlimit]
ir_steady = ir[np.logical_and(ir >= -irlimit, ir <= irlimit)]
ir_strong = ir[ir > irlimit]

eye_pass_count_list = []
for year in ['2021', '2022']:
    for date in md[year]['eyewall_limits']:
        # empty list case
        if len(md[year]['eyewall_limits'][date][0]) == 0:
            eye_pass_count_list.append(0)
        else:
            eye_pass_count_list.append(len(md[year]['eyewall_limits'][date]))
        # print(md[year]['eyewall_limits'][date])

# print out categories to be used
print("Dates:")
print( list(md['2021']['defined_eyewall'].keys()) + list(md['2022']['defined_eyewall'].keys()))
print("Defined eyewalls:")
print(defined_eyewalls)
print("Number of eye passes:")
print(eye_pass_count_list)
```

```python
# make nice intensity + intensification histograms here!
plt.figure(figsize=(12, 10))
bw1 = 5
bw2 = 5
bw3 = 2.5
bw4 = 30
helper_fns_winter2023.change_font_sizes(18, 18)

xinc1 = np.arange(0,150,bw1)
xinc2 = np.arange(-50,70,bw2)
xinc3 = np.arange(0,50,bw3)
xinc4 = np.arange(0,361,bw4)

# do this for each intensity category!
plt.subplot( 221)

ilist = [i_td, i_ts, i_wh, i_sh]
colors = ['g', 'c', 'y', 'r']
labels = ['TD', 'TS', 'WH', 'SH']
for i, ival in enumerate(ilist):
    hx=np.histogram( ival,xinc1)
    plt.rcParams["figure.figsize"] = [5,5]
    plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = colors[i], width = bw1, linewidth = 1, label=labels[i])
    plt.ylabel('Count');
    plt.xlabel('Intensity (kt)')
plt.legend(loc='upper right')
plt.text(x=55, y=4.4, s="TC Distributions: All Flights") #  for Cases with Clear CRL Eyewalls")

# repeat for intensification
plt.subplot( 222)
ilist = [ir_weak, ir_steady, ir_strong]
colors = ['k', 'b', 'r']
labels = ['Weakening', 'Steady State', 'Strengthening']
for i, ival in enumerate(ilist):
    hx=np.histogram( ival,xinc2)
    plt.rcParams["figure.figsize"] = [5,5]
    plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = colors[i], width = bw2, linewidth = 1, label=labels[i])
    plt.ylabel('Count');
    plt.xlabel('24 hr Intensification (kt)')    
plt.legend(loc='upper left', fontsize=13.5)

# repeat for shear strength
plt.subplot( 223)
hx=np.histogram( shearmag,xinc3)
plt.rcParams["figure.figsize"] = [5,5]
plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = 'Grey', width = bw3, linewidth = 1)
plt.ylabel('Count')
plt.xlabel('Shear Magnitude (kt)')    
plt.xlim([0, 30])

# repeat for shear direction
plt.subplot( 224)
hx=np.histogram( sheardir, xinc4)
plt.rcParams["figure.figsize"] = [5,5]
plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = 'lightblue', width = bw4, linewidth = 1)
plt.ylabel('Count')
plt.xlabel('Shear Direction (Degrees)')    
plt.axvline(x=90, c='k', linewidth=3, label='Westerly Winds')
plt.legend(loc='upper right', fontsize=13.5)
plt.xlim([0, 360])
```

```python
# Repeat of the plot shown above, but for total number of flight legs, not just flight days
# setup in this cell

eye_pass_count_list = []
for year in ['2021', '2022']:
    for date in md[year]['eyewall_limits']:
        # empty list case
        if len(md[year]['eyewall_limits'][date][0]) == 0:
            eye_pass_count_list.append(0)
        else:
            eye_pass_count_list.append(len(md[year]['eyewall_limits'][date]))
        # print(md[year]['eyewall_limits'][date])

# print out categories to be used
print("Dates:")
print( list(md['2021']['defined_eyewall'].keys()) + list(md['2022']['defined_eyewall'].keys()))
print("Defined eyewalls:")
print(defined_eyewalls)
print("Number of eye passes:")
print(eye_pass_count_list)

i_td, i_ts, i_wh, i_sh = [], [], [], []
ir_weak, ir_steady, ir_strong = [], [], []
shearmag, sheardir = [], []

count = 0
total_pass_number = 0
# repeat looping through dates, and sort by intensification here
for year in ['2021', '2022']:
    for date in md[year]['eyewall_limits']:
        passcount = eye_pass_count_list[count]
        
        # append the intensity, intensification, etc info multiple times for each pass
        # ex, if 2 passes, included intensity, etc readings 2 times
        for i in range(passcount):
            total_pass_number += 1
            
            intense = md[year]['intensity'][date]
            # add intensity info
            if intense < 34:
                i_td.append(intense)
            elif intense >= 34 and intense < 64:
                i_ts.append(intense)
            elif intense >= 64. and intense < 96.:                
                i_wh.append(intense)
            else:
                i_sh.append(intense)

            irate = md[year]['intensification'][date]
            # add intensification info
            irlimit = 10
            if irate < -irlimit:
                ir_weak.append(irate)
            elif irate >= -irlimit and irate <= irlimit:
                ir_steady.append(irate)
            else:
                ir_strong.append(irate)
                
            # add shear info
            shearmag.append(md[year]['shear_mag'][date])
            sheardir.append(md[year]['shear_dir'][date])
        count += 1
        
print("number of passes included: " + str(total_pass_number))
```

```python
# finally make the plots

plt.figure(figsize=(12, 10))
bw1 = 5
bw2 = 5
bw3 = 2.5
bw4 = 30
helper_fns_winter2023.change_font_sizes(18, 18)

xinc1 = np.arange(0,150,bw1)
xinc2 = np.arange(-50,70,bw2)
xinc3 = np.arange(0,50,bw3)
xinc4 = np.arange(0,361,bw4)

# do this for each intensity category!
plt.subplot( 221)

ilist = [i_td, i_ts, i_wh, i_sh]
colors = ['g', 'c', 'y', 'r']
labels = ['TD', 'TS', 'WH', 'SH']
for i, ival in enumerate(ilist):
    hx=np.histogram( ival,xinc1)
    plt.rcParams["figure.figsize"] = [5,5]
    plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = colors[i], width = bw1, linewidth = 1, label=labels[i])
    plt.ylabel('Count');
    plt.xlabel('Intensity (kt)')
plt.legend(loc='upper right')
plt.text(x=75, y=22, s="TC Distributions: All Eye Passes")
plt.yticks([0, 5, 10, 15, 20])

# repeat for intensification
plt.subplot( 222)
ilist = [ir_weak, ir_steady, ir_strong]
colors = ['k', 'b', 'r']
labels = ['Weakening', 'Steady State', 'Strengthening']
for i, ival in enumerate(ilist):
    hx=np.histogram( ival,xinc2)
    plt.rcParams["figure.figsize"] = [5,5]
    plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = colors[i], width = bw2, linewidth = 1, label=labels[i])
    plt.ylabel('Count');
    plt.xlabel('24 hr Intensification (kt)')    
plt.legend(loc='upper left', fontsize=13.5)
plt.yticks([0, 5, 10, 15, 20])

# repeat for shear strength
plt.subplot( 223)
hx=np.histogram( shearmag,xinc3)
plt.rcParams["figure.figsize"] = [5,5]
plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = 'Grey', width = bw3, linewidth = 1)
plt.ylabel('Count')
plt.xlabel('Shear Magnitude (kt)')    
plt.xlim([0, 30])
plt.yticks([0, 5, 10, 15, 20, 25])

# repeat for shear direction
plt.subplot( 224)
hx=np.histogram( sheardir, xinc4)
plt.rcParams["figure.figsize"] = [5,5]
plt.bar(hx[1][:-1],hx[0],edgecolor = 'k', color = 'lightblue', width = bw4, linewidth = 1)
plt.ylabel('Count')
plt.xlabel('Shear Direction (Degrees)')    
plt.axvline(x=90, c='k', linewidth=3, label='Westerly Winds')
plt.legend(loc='upper right', fontsize=13.5)
plt.xlim([0, 360])
plt.yticks([0, 5, 10, 15, 20, 25, 30])
```
