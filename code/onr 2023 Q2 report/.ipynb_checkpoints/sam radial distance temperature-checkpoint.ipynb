```python
import os
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import warnings

os.chdir("/Users/etmu9498/research/code/scripts-winter2023")
import helper_fns_winter2023 as helper_fns
```

```python
os.chdir("/Users/etmu9498/research/data/crl-all-data-processed/2021")
crl = xr.open_dataset("P3_20210926H1_processed.nc")

time = crl.time.values
dist = crl.center_dist.values
temp = crl.T.values
height = crl.height / 1000

xlims = [22.2, 25.8]

plt.figure(figsize=(14, 8))
helper_fns.change_font_sizes(14, 14)

plt.subplot(211)
plt.title("Improved Radial Distance Axis for TC Sam, 9/26/21")
plt.plot(time, dist, c='g')
helper_fns.add_blank_colorbar()
plt.ylim([0, 300])
plt.xlim(xlims)
plt.ylabel("Distance from TC Center (Km)")

plt.subplot(212)
warnings.filterwarnings("ignore")
color_map = plt.cm.get_cmap( "RdYlBu").reversed()
plt.pcolormesh( time, height, temp.transpose(), cmap = color_map, vmin=5, vmax=30 )
cbar = plt.colorbar()
cbar.ax.set_ylabel( ylabel="CRL Temperature ( C)", fontsize=20)
cbar.ax.tick_params(labelsize=20)
plt.ylabel( 'Height (km)')
plt.grid( 'on')
ax = plt.gca()
ax.set_facecolor('k')
warnings.filterwarnings("default")
plt.xlabel("Time (Hours, UTC)")
plt.xlim(xlims)
plt.ylim([0, 3.1])
```
