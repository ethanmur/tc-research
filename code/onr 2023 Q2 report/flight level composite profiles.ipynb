```python
# import...
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

os.chdir("/Users/etmu9498/research/code/scripts-winter2023/comps-ii-plot-helpers/sam-fl-comparisons")
import bin_rmw_data
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/ships-data-munging")
import group_ships_variables
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/")
import helper_fns_winter2023
```

```python
tc_df = group_ships_variables.group_all(tc='all', make_histograms=True, ri_cond=24)
```

```python
# find only the strong tcs (cat 3 or higher) that have just undergone rapid intensification
# make a new df
ri_cases = tc_df.iloc[ np.where( tc_df['Intensification (kt)'] >= 24.)]
strong_ri_cases = ri_cases.iloc[ np.where( ri_cases['Vmax (kt)'] >= 95.)]

# then, turn those text file names into inputs for the binning algorithm!
ri_names = []
ri_years = []
for namei, nameval in enumerate( strong_ri_cases['Name']):
    ri_names.append( nameval[:-4] + '_processed.nc')
    ri_years.append( nameval[0:4])

# print(strong_ri_cases)
# print(ri_names)
```

```python
# now, run the binning and plotting code on these cases! composite them :)

# make the input dictionary
inputdict = {}
for yeari, yearval in enumerate(ri_years):
    inputdict[yearval] = np.array(ri_names)[ np.where( np.array(ri_years) == yearval)[0]]

mb = 2.5
bw = .025
df = bin_rmw_data.bin(tc=inputdict, binwidth=bw, maxbin=mb)
print('data binned')
df = bin_rmw_data.plot_setup( df)
print('confidence intervals found')

# one color for each variable
# colors = [ 'y', 'r', 'k', 'Grey', 'g', 'b']
colors = [ 'b', 'b', 'b', 'b', 'b', 'b']


plt.figure( figsize=(22, 11))
lw = 2.
helper_fns_winter2023.change_font_sizes(18, 18)
subplots = [ 231, 232, 233, 234, 235, 236]


# new labels: fewer variables
ylabels = [ "Total Wind Speed (m/s)", "Temperature (C)", "WVMR (g/kg)", 'Vertical Winds (m/s)', 'Theta E (K)', 'Rain Rate (mm)'] 
var_names = ['WS.d', 'TA.d', 'MR.d', 'UWZ.d', 'THETAE.d', 'SfmrRainRate.1'] 


# add plots for every variable
for i in range( len( subplots)):
    plt.subplot( subplots[ i])
    plt.plot( df[ 'midpoints'], df[ var_names[ i] + '_mean'], c=colors[ i], linewidth=lw, label="RI Cases")
    ax = plt.gca()
    ax.fill_between( df[ 'midpoints'], df[ var_names[ i] + '_lowc'], df[ var_names[ i] + '_highc'],
                    color=colors[ i], alpha=0.3)

    if subplots[ i] >= 234:
        plt.xlabel("Radius of Maximum Wind")

    if subplots[i] == 234:
        plt.axhline(y=0., c='k')
        
    plt.xlim( [-.2, mb - .2])
    plt.ylabel( ylabels[ i])    

plt.text(-4, 74, "Flight Level Data Composites: RI and RW Cases for Cat. 3-5 TCs")



# find only the strong tcs (cat 3 or higher) that have just undergone rapid weakening... the other side of the story!
# make a new df
rw_cases = tc_df.iloc[ np.where( tc_df['Intensification (kt)'] <= -20.)]
strong_rw_cases = rw_cases.iloc[ np.where( rw_cases['Vmax (kt)'] >= 95.)]


# then, turn those text file names into inputs for the binning algorithm!
rw_names = []
rw_years = []
for namei, nameval in enumerate( strong_rw_cases['Name']):
    rw_names.append( nameval[:-4] + '_processed.nc')
    rw_years.append( nameval[0:4])

# now, run the binning and plotting code on these cases! composite them :)

# make the input dictionary
inputdict = {}
for yeari, yearval in enumerate(rw_years):
    inputdict[yearval] = np.array(rw_names)[ np.where( np.array(rw_years) == yearval)[0]]

mb = 2.5
bw = .025
df = bin_rmw_data.bin(tc=inputdict, binwidth=bw, maxbin=mb)
print('data binned')
df = bin_rmw_data.plot_setup( df)
print('confidence intervals found')

# one color for each variable
# colors = [ 'y', 'r', 'k', 'Grey', 'g', 'b']
colors = ['r', 'r', 'r', 'r', 'r', 'r']

# plt.figure( figsize=(21, 12))
lw = 2.
helper_fns_winter2023.change_font_sizes(18, 18)
subplots = [ 231, 232, 233, 234, 235, 236]


# new labels: fewer variables
ylabels = [ "Total Wind Speed (m/s)", "Temperature (C)", "WVMR (g/kg)", 'Vertical Winds (m/s)', 'Theta E (K)', 'Rain Rate (mm)'] 
var_names = ['WS.d', 'TA.d', 'MR.d', 'UWZ.d', 'THETAE.d', 'SfmrRainRate.1'] 


# add plots for every variable
for i in range( len( subplots)):
    plt.subplot( subplots[ i])
    plt.plot( df[ 'midpoints'], df[ var_names[ i] + '_mean'], c=colors[ i], linewidth=lw, label = "RW Cases")
    ax = plt.gca()
    ax.fill_between( df[ 'midpoints'], df[ var_names[ i] + '_lowc'], df[ var_names[ i] + '_highc'],
                    color=colors[ i], alpha=0.3)

    if subplots[ i] >= 234:
        plt.xlabel("Radius of Maximum Wind")

    if subplots[i] == 231:
        plt.legend(loc='lower right')
        
    if subplots[i] == 234:
        plt.axhline(y=0., c='k')

    plt.xlim( [-.2, mb - .2])
    plt.ylabel( ylabels[ i])    
```
