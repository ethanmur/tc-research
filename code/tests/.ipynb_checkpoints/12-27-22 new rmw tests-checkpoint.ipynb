```python
# setup
tcname = 'sam'
dataset = 0
metadata = tc_metadata.all_data( tcname)

# load data
tdr_name, crl_name = tc_metadata.choose_new_data( tcname, dataset)
crl_path =  "/Users/etmu9498/research/data/crl-new-matrices"
os.chdir( crl_path)
crl_data = xr.open_dataset( crl_name)

flight_name = tc_metadata.choose_new_in_situ_name( tcname, dataset)
flight_path = "/Users/etmu9498/research/data/in-situ-new"
os.chdir( flight_path)
new_crl = xr.open_dataset( flight_name)
```

```python
# check that eyewall lims and data names are correct
print( crl_name)
print( flight_name)
print( metadata['eyewall_dists_no_eyewalls'][dataset])
```

```python
# find the center point of the eye: equal number of data points to the left as to the right

lim0 = metadata['eyewall_dists_no_eyewalls'][dataset][0]
lim1 = metadata['eyewall_dists_no_eyewalls'][dataset][1]
dist = crl_data.in_situ_distance
# the indices closest to lim0 and lim1
ind0 = (np.abs(crl_data.in_situ_distance - lim0 )).argmin().values
ind1 = (np.abs(crl_data.in_situ_distance - lim1 )).argmin().values

print( ind0)
print( ind1)

# trim down distance axis to look at the eye
trimdist = dist[ np.where( dist.values >= lim0)]
trimdist = trimdist[ np.where( trimdist.values <= lim1)]

# np.set_printoptions(threshold=np.inf)
# print( trimdist.values)
# np.set_printoptions(threshold=1000)
# print( len( trimdist.values))

# find the eye center
# even case
if len( trimdist.values) % 2 == 0:
    small_ind = len( trimdist.values) / 2
else:
    # math.ceil rounds the division up to get the center int!
    small_ind = math.ceil( len( trimdist.values) / 2)
    
# add all the indices before the first eyewall to get a properly spaced tc center
startdist = dist[ np.where( dist.values <= lim0)]
center_ind = int( len( startdist.values) + small_ind )
center_dist = dist[ center_ind]

# test that this operation works!
print( len( dist[ np.where( dist.values <= lim0)]))
# print( small_ind)
# print( center_ind)
```

```python
# testing: distance axes from each eyewall to tc center
firsthalf = crl_data.in_situ_distance[ ind0:center_ind]
secondhalf = crl_data.in_situ_distance[ center_ind:ind1]

# calculate the axis step: are they equal intervals?
# if not, there might be complications with using a fixed interval through linspace() for rmw generation
# for i in range( len( firsthalf) - 1):
# print( firsthalf[i].values - firsthalf[i+1].values)


rmwminus = np.linspace( -1, 0, num=len( firsthalf)).tolist()
rmwplus = np.linspace( 0, 1, num=len( secondhalf)).tolist()[1:] # [1:] removes the first value, a repeat 0.0
    
# actually this doesn't work... center isn't = 0 here
# divide by largest number to normalize to 1
# rmwplus = secondhalf / secondhalf[ -1]

# concatenate lists to make one new rmw axis!
# rmwaxis = rmwminus + rmwplus


##########################
## simpler and probably better way of doing things!! just initialize both sides of the rmw at once
##########################

rmwaxis = np.linspace( -1, 1, num=len(firsthalf)+len(secondhalf))


# find inds closest to rmw = +-.5
# add distance to ind = 0 for eyewall like above!
ind_minus5 = (np.abs(rmwaxis + .5 )).argmin() + len( startdist.values)
ind_plus5 = (np.abs(rmwaxis - .5 )).argmin() + len( startdist.values)

print( ind_minus5)
print( ind_plus5)

# print( rmwminus)
# print( rmwplus)
print( rmwaxis)
# print( len( rmwaxis))
```

```python
# find cloud heights for current eye
H, xaxis_value = cloud_height.find_cloud_heights( crl_name, -30, ind0, ind1, xaxis ='in-situ-dist', crl_path=crl_path, new_heights=True)

# print( len( H))
# print( H)
# print( xaxis_value)

# break the distribution down by rmw distance!
lowrmw = []
highrmw = []
# cycle through every case
for i in range( len( H)):
    
    # high rmw case: add cloud heights
    if rmwaxis[ i] > .5 or rmwaxis[ i] < -.5:
        highrmw.append( H[ i])
    # low rmw case
    else:
        lowrmw.append( H[ i])
        
print( len( lowrmw))
print( len( highrmw))
print( lowrmw)
print( highrmw)
```

```python
# plot simple dists of both cases: outer and inner eye clouds!
plt.figure(figsize=(14, 7))
sns.set_theme(style="white", palette=None)
helper_fns.change_font_sizes()

# the line below was probs causing the annoying lack of plot space!!!
# sns.set_context(rc = {'patch.linewidth': 0.0})
plt.subplot(121)
sns.histplot( y= lowrmw, kde=True, binwidth=.175, edgecolor='k', linewidth=2, color = 'g') # color = 'g', element='poly') # hist_kws=dict(edgecolor="black", linewidth=2))
plt.xlabel( 'Count for Each Cloud Height')
plt.ylabel( 'Height from Surface (Km)')
plt.title( "Cloud Height Distribution for Inner Eye Clouds")
plt.ylim( [-.2, 3.75])
plt.xlim( [0, 30])
plt.grid(True)

plt.subplot(122)
sns.histplot( y= highrmw, kde=True, binwidth=.175, edgecolor='k', linewidth=2, color = 'g') # color = 'g', element='poly') # hist_kws=dict(edgecolor="black", linewidth=2))
plt.xlabel( 'Count for Each Cloud Height')
plt.ylabel( 'Height from Surface (Km)')
plt.title( "Cloud Height Distribution for Outer Eye Clouds")
plt.ylim( [-.2, 3.75])
plt.xlim( [0, 30])
plt.grid(True)
```
