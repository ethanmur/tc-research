```python
# some simple code to make a whisker plot showing:
# 8 mean data points with 75% and 90% confidence intervals
# 2 points for each of the 4 TC intensities, one for inner clouds and another for outer
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import scipy

os.chdir(  "/Users/etmu9498/research/code/scripts/plotting")
import rmw_inner_outer_stats as stats
os.chdir(  "/Users/etmu9498/research/code/scripts/")
import helper_fns
```

```python
heights, inner, outer = stats.plot_all(tc='all', test_type='return')

# find heights above 50m for every case, after running script above
cliph = []
clipinner = []
clipouter = []
for i in range( 4):    
    only_cloud_heights = np.array( heights[i]) [ np.where( np.array( heights[ i]) > .05)[0] ]
    cliph.append( only_cloud_heights)

    only_inner = np.array( inner[i]) [ np.where( np.array( inner[ i]) > .05)[0] ]
    clipinner.append( only_inner)

    only_outer = np.array( outer[i]) [ np.where( np.array( outer[ i]) > .05)[0] ]
    clipouter.append( only_outer)

print( len( heights))
print( len( inner))
print( len( outer))

print( len( heights[0]))
print( len( cliph[0]))
print( len( inner[0]))
print( len( outer[0]))
```

```python
# make a simple whisker plot for each distribution!
plt.figure( figsize=(16, 14))

helper_fns.change_font_sizes(18, 18)

plt.subplot(311)
# plt.xlabel( "TC Intensity Category")
plt.ylabel( "Cloud Height (km)")
plt.title( "Cloud Height Distributions: All Eye Clouds")

# locally define important variables
totalh = cliph

names = ["TD", "TS", "WH", "SH"]
xs = []

# add a "jitter" to each data point to avoid overlapping data
for j in range( len( totalh)):
    xs.append(np.random.normal(j + 1, 0.1, len( totalh[j]) ))

plt.boxplot( totalh, labels=names)

palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, totalh, palette):

    print( len( x))
    print( len( val))
    plt.scatter(x, val, alpha=0.4, color=c)
    
    
    
plt.subplot(312)
# plt.xlabel( "TC Intensity Category")
plt.ylabel( "Cloud Height (km)")
plt.title( "Inner Eye Clouds")

# locally define important variables
totalh = clipinner

names = ["TD", "TS", "WH", "SH"]
xs = []

# add a "jitter" to each data point to avoid overlapping data
for j in range( len( totalh)):
    xs.append(np.random.normal(j + 1, 0.1, len( totalh[j]) ))

plt.boxplot( totalh, labels=names)

palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, totalh, palette):

    print( len( x))
    print( len( val))
    plt.scatter(x, val, alpha=0.4, color=c)
    
    
    
plt.subplot(313)
plt.xlabel( "TC Intensity Category")
plt.ylabel( "Cloud Height (km)")
plt.title( "Outer Eye Clouds")


# locally define important variables
totalh = clipouter

names = ["TD", "TS", "WH", "SH"]
xs = []

# add a "jitter" to each data point to avoid overlapping data
for j in range( len( totalh)):
    xs.append(np.random.normal(j + 1, 0.1, len( totalh[j]) ))

plt.boxplot( totalh, labels=names)

palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, totalh, palette):

    print( len( x))
    print( len( val))
    plt.scatter(x, val, alpha=0.4, color=c)
    
```
