```python
# helper function used to turn string w values into floats
def in_situ_to_float( return_var):
    return_var_temp = np.zeros( len( return_var))
    for line_ind in range( len( return_var)):
        if return_var[ line_ind] == '':
            return_var_temp[line_ind] = np.nan
        else:
            return_var_temp[ line_ind] = float( return_var[ line_ind])
    return return_var_temp
```

```python
# taken from 'in-situ-scripts/save_in_situ_as_nedcdf.py'
def float_creater2( dataset):
    # this snippet turns all relevant quantities from strings into floats for easier use later
    keylist = [ 'time', 'WS.d', 'WD.d', 'UWZ.d', 'ASfmrRainRate.1', 'LATref', 'LONref', 'TAS.d', 'PSURF.d', 
               'HT.d', 'PITCHref', 'ROLLref', 'MR.d', 'HUM_REL.d', 'THETA.d', 'THETAE.d', 'THETAV.d', 'TA.d']
    # create an empty version of the xarray dataset
    datasetnew = dataset.copy()
    datasetnew = datasetnew[ []]

    for key in keylist:
        datasetnew[ key] = dataset[ key].astype( float)
    return datasetnew
```

```python
# test: smooth wind speeds and surface pressures
window = 30 # seconds

ws_avg = pd.Series( ws).rolling(window=window, min_periods=1, center=True).mean()
psurf_avg = pd.Series( psurf).rolling(window=window, min_periods=1, center=True).mean()

w_avg = pd.Series( w).rolling(window=window, min_periods=1, center=True).mean()
```

```python
# try to pick out minimum pressure values using raw and smoothed data: any differences?

# ways to find pressure minima:
# 1) use a threshold depending on the strnegth of the tc (strength can be found by finding the max ws!) to sort for only
# certain pressure minima ( less than 960 hPa for SH, less than 990 hPa for TS, etc). This will eliminate false peaks?

# 2) use some find_peaks trickery to isolate the correct central pressure peaks! Again, this might require filtering based
# on strength

# 3) use windows determined by crl data, and just look for the lowest pressure over that window -> simpler, but it 
# is significantly slower :(

# 4) use a self organizing map or another fancy machine learning algorithm to sort and find tc centers???
```

```python
# test 1: pressure threshold with no eyewall limits

# find max TC strength from averaged wind fields
vmax = np.nanmax( ws_avg)
print( vmax)

# check based on max TC strength
if vmax > 50:
    print( "SH case")
    peak_min = 40 # m/s

else:
    peak_min = 20
    
    
# find max peaks using the min peak height for this intensity
vpeaks = find_peaks( x=ws_avg, height=peak_min, prominence=10)

print( vpeaks)
# print( fl_data.time[ vpeaks[0][0]].values)
# print( vpeaks[1].values)
```

```python
# find the wind speed (and pressure I guess!) minima between the eyewall's max points!
ws_mins = []
p_mins = []
ws_time = []
p_time = []

ws_rmw = []
wtrim = []
vtrim = []

# do step = 2 to skip the second index for each eyewall!
for vind in range( int( len(vpeaks[0]) / 2 ) ):
    
    left_eyewall_ind = vpeaks[0][ 2* vind]
    right_eyewall_ind = vpeaks[0][2 * vind + 1]
    
    # find the minimum values within each eyewall limit
    ws_mins.append( np.nanmin( ws_avg[ left_eyewall_ind:right_eyewall_ind]))
    p_mins.append( np.nanmin( psurf_avg[ left_eyewall_ind:right_eyewall_ind]))
    
    # find the corresponding times
    ws_ind = np.nanargmin( ws_avg[ left_eyewall_ind:right_eyewall_ind])
    ws_time.append( fl_data.time[ left_eyewall_ind + ws_ind ].values )
    p_ind = np.nanargmin( psurf_avg[ left_eyewall_ind:right_eyewall_ind])
    p_time.append( fl_data.time[ left_eyewall_ind + p_ind ].values )
    
    # make an rmw axis stretching from one eyewall to the other!
    # they will have an unequal number of data points to conserve the time axis :)
    len1 = ws_ind
    len2 = right_eyewall_ind - ( left_eyewall_ind + ws_ind)
    ws_rmw.append( np.concatenate( [ np.linspace( -1, 0, num=len1, endpoint=False), np.linspace( 0, 1, num=len2) ] ))
   
    # add the v and w values for convenience
    vtrim.append( ws[ left_eyewall_ind : right_eyewall_ind])
    wtrim.append( w[ left_eyewall_ind : right_eyewall_ind])
    
# print( ws_mins)
# print( p_mins)
# print( p_time)
# print( ws_time)

print( len( ws_rmw[0]))
print( len( vtrim[0]))

print( ws_rmw[0])
print( vtrim[0])
```

```python
# make plots of total wind speed and vertical vels along new rmw axes!
plt.figure( figsize=(14, 11))
lw = 2.5
helper_fns.change_font_sizes(16, 16)

plt.subplot(211)

plt.title( rmwtitle)

plt.ylabel("Total Wind Speed (m/s)")
plt.xlim( [-1.2, 1.2])
    
    
# plot each of the vel profiles
for i in range( len( ws_rmw)):
    plt.plot( ws_rmw[i], vtrim[i], c=colors[i], label=labels[i], linewidth=lw)
plt.legend(loc='lower left')
plt.grid()

plt.subplot(212)
plt.ylabel("Vertical Velocity (m/s)")
plt.xlabel("Radius of Maximum Wind")
plt.xlim( [-1.2, 1.2])

# plot each of the vel profiles
for i in range( len( ws_rmw)):
    plt.plot( ws_rmw[i], wtrim[i], c=colors[i], label=labels[i], linewidth=lw)
plt.legend(loc='lower left')
plt.grid()
```
