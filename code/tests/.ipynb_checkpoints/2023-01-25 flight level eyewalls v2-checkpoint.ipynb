```python
# import...
import numpy as np
import xarray as xr
import os
import matplotlib.pyplot as plt
from matplotlib import cm

import warnings
import pandas as pd
from scipy.signal import find_peaks

os.chdir("/Users/etmu9498/research/code/scripts")
import make_plots_new_heights
import make_plots
import tc_metadata
import helper_fns

os.chdir("/Users/etmu9498/research/code/scripts/in-situ-scripts")
import load_in_situ_data

os.chdir("/Users/etmu9498/research/code/scripts/plotting")
import rmw_inner_outer_stats as extra_stats # confidence interval function resides here!
```

```python
# taken from 'in-situ-scripts/save_in_situ_as_nedcdf.py'
# old function is way too clunky and slow, this one works much better!

def tofloat( dataset):
    # this snippet turns all relevant quantities from strings into floats
    # add more fields here if needed!
    keylist = [ 'time', 'WS.d', 'WD.d', 'UWZ.d', 'ASfmrRainRate.1', 'LATref', 'LONref', 'TAS.d', 'PSURF.d', 
               'HT.d', 'PITCHref', 'ROLLref', 'MR.d', 'HUM_REL.d', 'THETA.d', 'THETAE.d', 'THETAV.d', 'TA.d']
    # create an empty version of the xarray dataset
    datasetnew = dataset.copy()
    datasetnew = datasetnew[ []]

    for key in keylist:
        datasetnew[ key] = dataset[ key].astype( float)
    return datasetnew
```

```python
######################
## Load and find rmw values for each TC!
######################

# load 2021 in situ data
fl_path = "/Users/etmu9498/research/data/in-situ/"
fl_list = make_plots.load_flight_level( fl_path, print_files=False)

# for this notebook, do this for every case from TC Sam! 
dataset_numbers = [29, 31, 32, 34] # 20, this final dataset looks really bad :( excluded from this analysis! 35]
dates = ["09-25", "09-26", "09-27", "09-29"] # '08-29' , "09-30"]
names = ['Sam', 'Sam', 'Sam', 'Sam'] # 'Ida'

# lists to hold total / vertical vels + x axes values
# values are nested as a list of lists, like: [ [day 1], [day 2], ...], where [day2] = [ [pass 1], [pass 2], ...]
t_total = []
rmw_total = []
w_total = []
spd_total = []

# new variables! 
temp_total = []
thetav_total = []
wv_total = []
rr_total = [] # rain rate
psurf = [] # surface pressure

########################
# maybe make another loop below to iterate through the variables?? Adding vars individually is getting tedius lol
########################

pass_counter = 0

# add all total and vertical vels to one plot! within rmw
for i in range( len( dataset_numbers)):
    data_number = dataset_numbers[ i]
    fl_txt = fl_list[ data_number] # original case: 31
    fl_data = load_in_situ_data.load_in_situ( fl_path, fl_txt)
    dataset = tofloat( fl_data)
    print( "Dataset " + str( i) + " loaded")
                

    # smooth wind speeds and surface pressures- more realistic peaks
    window = 30 # seconds
    spd_avg = pd.Series( dataset['WS.d']).rolling(window=window, min_periods=1, center=True).mean()
    psurf_avg = pd.Series( dataset['PSURF.d']).rolling(window=window, min_periods=1, center=True).mean()
    w_avg = pd.Series( dataset['UWZ.d']).rolling(window=window, min_periods=1, center=True).mean()
    
    # test 1: pressure threshold with no eyewall limits
    # find max TC strength from averaged wind fields- all eye passes from this tc
    vmax = np.nanmax( spd_avg)
    # check based on max TC strength
    if vmax > 50:
        print( "SH case")
        peak_min = 40 # m/s
    else: 
        print( "Weaker case: update if statement for different peak_min behavior!")
        peak_min = 40 # m/s

    # find max vel peaks using the min peak height (defined above) for this intensity
    vpeaks = find_peaks( x=spd_avg, height=peak_min, prominence=10)

    
    
    # find the wind speed (and pressure I guess!) minima between the eyewall's max points!
    # ws_mins = []
    # p_mins = []
    # p_time = []

    ws_time = []
    ws_rmw = []
    wtrim = []
    vtrim = []
    
    temp_trim = []
    thetav_trim = []
    wv_trim = []
    rr_trim = []

    # do step = 2 to skip the second index for each eyewall!
    for vind in range( int( len(vpeaks[0]) / 2 ) ):

        left_eyewall_ind = vpeaks[0][ 2* vind]
        right_eyewall_ind = vpeaks[0][2 * vind + 1]

        # find the minimum values within each eyewall limit
        # ws_mins.append( np.nanmin( spd_avg[ left_eyewall_ind:right_eyewall_ind]))
        # p_mins.append( np.nanmin( psurf_avg[ left_eyewall_ind:right_eyewall_ind]))

        # find the corresponding times
        ws_ind = np.nanargmin( spd_avg[ left_eyewall_ind:right_eyewall_ind])
        ws_time.append( fl_data.time[ left_eyewall_ind + ws_ind ].values )
        # p_ind = np.nanargmin( psurf_avg[ left_eyewall_ind:right_eyewall_ind])
        # p_time.append( fl_data.time[ left_eyewall_ind + p_ind ].values )

        # make an rmw axis stretching from one eyewall to the other!
        # they will have an unequal number of data points to conserve the time axis :)
        len1 = ws_ind
        len2 = right_eyewall_ind - ( left_eyewall_ind + ws_ind)
        ws_rmw.append( np.concatenate( [ np.linspace( -1, 0, num=len1, endpoint=False), np.linspace( 0, 1, num=len2) ] ))

        # add the v and w values for convenience
        vtrim.append( dataset["WS.d"][ left_eyewall_ind : right_eyewall_ind].values)
        wtrim.append( dataset["UWZ.d"][ left_eyewall_ind : right_eyewall_ind].values)

        # new variables 
        temp_trim.append( dataset["TA.d"][ left_eyewall_ind : right_eyewall_ind].values)
        thetav_trim.append( dataset["THETAE.d"][ left_eyewall_ind : right_eyewall_ind].values)
        wv_trim.append( dataset["MR.d"][ left_eyewall_ind : right_eyewall_ind].values)
        rr_trim.append( dataset["ASfmrRainRate.1"][ left_eyewall_ind : right_eyewall_ind].values)

        pass_counter += 1

    # new step: append lists from current flight date to the total lists!
    t_total.append( ws_time)
    rmw_total.append( ws_rmw)
    w_total.append( wtrim)
    spd_total.append( vtrim)
    
    temp_total.append( temp_trim)
    thetav_total.append( thetav_trim)
    wv_total.append( wv_trim)
    rr_total.append( rr_trim)
```

```python
# pick a bin width for rmw values, sort velocities into those bins, and 
# calculate important quantities like the mean, standard deviation, and confidence levels! 

### important! instead of going from -1 < RMW < 1, sort from 0 to 1! the negative distinction doesn't matter here
# setup steps

# .005 is an interesting bin width! anything larger than .01 lacks resolution, smaller than .005 is too noisy
binwidth = .01 # bin width from 0 to 1

# make an array representing the bins! 
bins = np.arange(0, 1.00001, step=binwidth)

# create an array of midpoints (right between two given binwidths)
# will be 1 val shorter than bins array
# used for plotting later
midpoints = []
for i in range( len( bins) - 1):
    midpoints.append( (bins[ i] + bins[ i+1]) / 2)

# print( bins)
# print( midpoints)

# initialize a list of lists, number of inner lists should be = number of bins
rmwsorted = []
spdsorted = []
wsorted = []

# new vars
tempsorted = []
thetavsorted = []
wvsorted = []
rrsorted = []

for i in range( len( bins) - 1):
    rmwsorted.append( [])
    spdsorted.append( [])
    wsorted.append( [])

    tempsorted.append( [])
    thetavsorted.append( [])
    wvsorted.append( [])
    rrsorted.append( [])

# print( len( spdsorted))
# print( len( midpoints))

# simple sum to print the current pass number for the user
case_count = 0
```

```python
# make plots like the ones above, but with the mean values and confidence intervals instead of individual passes!

# average each of the speeds saved in the separate arrays (all the speeds belonging to that bin)
spdavg = []
spdlowc = []
spdhighc = []

wavg = []
wlowc = []
whighc = []

tempavg = []
templowc = []
temphighc = []

thetavavg = []
thetavlowc = []
thetavhighc = []

rravg = []
rrlowc = []
rrhighc = []

wvavg = []
wvlowc = []
wvhighc = []


confidence=.95
for i in range( len( midpoints)):
    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( spdsorted[ i], np.isnan( spdsorted[ i])), confidence=confidence)
    spdavg.append(  mean)
    spdlowc.append(  lowc)
    spdhighc.append(  highc)

    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( wsorted[ i], np.isnan( wsorted[ i])), confidence=.95)
    wavg.append(  mean)
    wlowc.append(  lowc)
    whighc.append(  highc)
    
    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( tempsorted[ i], np.isnan( tempsorted[ i])), confidence=.95)
    tempavg.append(  mean)
    templowc.append(  lowc)
    temphighc.append(  highc)
    
    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( thetavsorted[ i], np.isnan( thetavsorted[ i])), confidence=.95)
    thetavavg.append(  mean)
    thetavlowc.append(  lowc)
    thetavhighc.append(  highc)
    
    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( rrsorted[ i], np.isnan( rrsorted[ i])), confidence=.95)
    rravg.append(  mean)
    rrlowc.append(  lowc)
    rrhighc.append(  highc)
    
    mean, lowc, highc = extra_stats.mean_confidence_interval( 
                np.ma.masked_array( wvsorted[ i], np.isnan( wvsorted[ i])), confidence=.95)
    wvavg.append(  mean)
    wvlowc.append(  lowc)
    wvhighc.append(  highc)
```
