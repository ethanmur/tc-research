```python
# make a simple plot to compare distance axis from "2023-02-15 flight tracks to radial dist v3"

# these plots look good!

plt.figure( figsize=(14, 5))

# make a simple plot showing pressure drops overlapping with distance from the tc center
lw = 2
helper_fns.change_font_sizes(16, 16)

time = data['float_time']
p = data['PSURF.d']
dist = data["center_dist"]
rmw = data[ "rmw"]

ax = plt.gca()
ax.plot( time, p, c='k', linewidth=lw)
ax.set_ylabel( "P Surface (hPa)")
ax.set_title( "Pressure and Radial Distance Relationships")
ax.set_xlabel("Time (Hours, UTC)")

ax2 = ax.twinx()
ax2.plot( time, rmw, c='r', linewidth=lw)

# ax2.set_ylabel( "Distance from Center (Km)")

# ax2.set_ylim( [0, 500])
# ax2.set_ylim( [0, 50])
plt.grid()
```

```python
# print out all rmw axes recursively for all 2021 data! see how they look

datadir = "/Users/etmu9498/research/data/in-situ-noaa-processed/2021"
files = make_plots.load_flight_level( datadir, print_files=False)

print( files)

for filei, fileval in enumerate( files):

    os.chdir( datadir)
    data = xr.open_dataset( fileval, decode_times=False)
    
    plt.figure( figsize=(14, 10))

    # make a simple plot showing pressure drops overlapping with distance from the tc center
    lw = 2
    helper_fns.change_font_sizes(16, 16)

    time = data['float_time']
    p = data['PSURF.d']
    dist = data["center_dist"]
    rmw = data[ "rmw"]

    plt.subplot(211)
    ax = plt.gca()
    ax.plot( time, p, c='k', linewidth=lw)
    ax.set_ylabel( "P Surface (hPa)")
    ax.set_title( "Pressure and Radial Distance Relationships: " + fileval)
    xlims = ax.get_xlim()

    ax2 = ax.twinx()
    ax2.plot( time, rmw, c='r', linewidth=lw)
    ax2.set_ylabel( "RMW")
    plt.grid()

    plt.subplot( 212)
    plt.plot( time, dist, c='b', linewidth=lw)
    plt.ylabel( "Radial Distance (km)")
    plt.xlabel("Time (Hours, UTC)")
    plt.xlim( xlims)
    plt.grid()
```
