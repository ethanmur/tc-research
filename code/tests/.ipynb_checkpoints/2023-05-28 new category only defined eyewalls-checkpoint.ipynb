```python
import os
import xarray as xr
import numpy as np

os.chdir("/Users/etmu9498/research/code/scripts-winter2023/")
import helper_fns_winter2023 as helper_fns
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/cloud-top-height-stats")
import eyewall_metadata
```

```python
# load metadata
metadata = eyewall_metadata.all_metadata()
print(metadata['2021'].keys())
```

```python
# print out basic metadata info for each tc date

# do this for both years
for year in metadata.keys():
    print("Year " + year)
    # do this for every date
    for date in metadata[year]['names']:
        print("Category for " + date + ": " + str(metadata[year]['category'][date]))
        print("Number of Eye Passes: " + str( len( metadata[year]['eyewall_limits'][date])))
        print("Defined eyewall? " + str(metadata[year]['defined_eyewall'][date]))
        print("Intensification (kt): " + str(metadata[year]['intensification'][date]) + "\n")
```

```python
# print out the number of dates and eye passes for defined eyewalls- per year and total
all_date_count = 0
all_pass_count = 0

print("Defined eyewalls")
print("Note: the defined eye passes sum includes all cases with one valid eye! Maybe a slight over estimate")

# do this for both years
for year in metadata.keys():
    year_date_count = 0
    year_pass_count = 0
    print("Year " + year)
    
    # do this for every date
    for date in metadata[year]['names']:
        # only add counts if there's a defined eyewall
        if metadata[year]['defined_eyewall'][date]:
            all_date_count += 1
            all_pass_count += len( metadata[year]['eyewall_limits'][date])

            year_date_count += 1
            year_pass_count += len( metadata[year]['eyewall_limits'][date])

    # print out stats for this year
    print("Number of Defined Eye Dates: " + str(year_date_count))
    print("Number of Defined Eye Passes: " + str(year_pass_count))
            
# print out total counts
print("Total")
print("Number of Defined Eye Dates: " + str(all_date_count))
print("Number of Defined Eye Passes: " + str(all_pass_count))
```

```python
# Repeat for undefined eyewalls!
all_date_count = 0
all_pass_count = 0

print("Undefined eyewalls")
print("note: the date sum includes the 4 crl datasets that don't have a single valid eye pass!")

# do this for both years
for year in metadata.keys():
    year_date_count = 0
    year_pass_count = 0
    print("Year " + year)
    
    # do this for every date
    for date in metadata[year]['names']:
        
        # only add counts if there's a defined eyewall
        if not metadata[year]['defined_eyewall'][date]:
            all_date_count += 1
            all_pass_count += len( metadata[year]['eyewall_limits'][date])

            year_date_count += 1
            year_pass_count += len( metadata[year]['eyewall_limits'][date])

    # print out stats for this year
    print("Number of Undefined Eyewall Dates: " + str(year_date_count))
    print("Number of Undefined Eyewall Passes: " + str(year_pass_count))
            
# print out total counts
print("Total")
print("Number of undefined Eyewall Dates: " + str(all_date_count))
print("Number of undefined Eyewall Passes: " + str(all_pass_count))
```
