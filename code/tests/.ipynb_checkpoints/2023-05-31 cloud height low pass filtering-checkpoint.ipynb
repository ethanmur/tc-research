```python
# goal: test the effects of applying a low pass filter to cloud height data! Does it lead to any noticable improvements
# in peak detection, etc?

# results: the butterworth filter works fairly well! the really scattered heights can be smoothed effectively.
# the find_peaks filter is good at identifying convective cells, but not so good at finding flat stratus decks.
# so, not a good cloud cluster algorithm...
# BUT, this can be used to find convective cells vs stratiform cells! find where a cluster overlaps with a peak -> 
# convection
```

```python
# import...
import os
import sys
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal as signal

sys.path.append("/Users/etmu9498/research/code/scripts-winter2023")
import helper_fns_winter2023
sys.path.append("/Users/etmu9498/research/code/scripts-winter2023/cloud-top-height-stats")
import eyewall_metadata
import find_cloud_tops
```

```python
# this code was mostly taken from cloud_top_plotting_clusters.py

# get lidar data for this specific day
metadata = eyewall_metadata.all_metadata()
lw = 2
fs = 20

crl_path = "/Users/etmu9498/research/data/crl-all-data-processed/2021/"
os.chdir( crl_path)
yearval = '2021'
fileval = "P3_20210812H2_processed.nc" # "P3_20210819H1_processed.nc"
casei = 0

crl_data = xr.open_dataset( fileval)

######
## new code: grab the limits for this case
######
date = fileval[7:11]

# check if this date exists... if not, give it some empty eyewall limits!
# also check for fred am and pm cases!!
if date == '0812':
    if fileval[11:13] == "H1":
        eyewall_limits = metadata[ yearval]['eyewall_limits'][ '0812am']
    elif fileval[11:13] == "H2":
        eyewall_limits = metadata[ yearval]['eyewall_limits'][ '0812pm']
elif date in metadata[ yearval]['eyewall_limits'].keys():
    eyewall_limits = metadata[ yearval]['eyewall_limits'][ date]
else:
    eyewall_limits = [ ()]

print("Date: " + date)
# make plots for proposed eyewall limits!

pairval = eyewall_limits[casei]

print( "Eyewall limits: " + str( pairval))
# only do this if there are eyewall limits to look at
if len( pairval) > 1:
    lefteyewall = pairval[ 0]
    righteyewall = pairval[ 1]

######################
# plot power ch1
# also find and plot cloud top heights and clusters!!
if yearval == '2021':
    min = -30
elif yearval == '2022':
    min = -40
max=-10

# preparation
li = np.argmin( np.abs(crl_data.time.values - lefteyewall))
ri = np.argmin( np.abs(crl_data.time.values - righteyewall))
H = crl_data.height
power = crl_data.P_ch1[ li:ri, :]
axis = crl_data.time[ li:ri]
p3_height = crl_data.p3_height[ li:ri]

# get cloud heights and clusters
cloudheights, cloudtime = find_cloud_tops.find_cloud_heights( H, power, axis, p3_height, cutoff_power = min)

# get the mean cloud height: useful for determining convective towers later!
meanh = np.nanmean(cloudheights)
print("mean height (m) = " + str(meanh))
```

```python
# smooth the cloud tops here!

### Apply a Butterworth filter (recursive filter) to the data
N = 9 ## order
frequency_cutoff = .20 # 0.15 # remove f's greater than .05 year^-1 (20 years)
Wn = frequency_cutoff*2 ## scalar given the critical frequency (all higher frequencies are removed)

## Note: Wn is normalized from 0 to 1, where 1 is the Nyquist frequency, pi radians/sample. 
## Note: (Wn is thus in half-cycles / sample.) 
b, a = signal.butter(N, Wn)
newcloudh = signal.filtfilt(b, a, cloudheights) ## one filter forward, one filter backward - you are filtering twice

# make sure filtered clouds don't dip below 0m
newcloudh[np.where(newcloudh < 0.0)[0]] = 0.0

# find peaks in this signal
# test 1: lower peaks with large prominence
peakinds1 = signal.find_peaks(x=newcloudh, prominence=500, height=meanh)[0] # width=3)[0]
# test 2: taller peaks with smaller prominence
peakinds2 = signal.find_peaks(x=newcloudh, prominence=100, height=meanh + meanh / 3)[0] # width=3)[0]
# join the two tests!
peakinds = np.union1d( peakinds1, peakinds2)

peakx, peakh = cloudtime[peakinds], newcloudh[peakinds]
```

```python
################
# make the figure!
plt.figure(figsize=(10, 5))
helper_fns_winter2023.change_font_sizes(fs, fs)
plt.title(str(date[0:2] + '/' + date[2:4] + ", Pass " + str(pairi) + " Eye Profile"))


p = plt.pcolormesh( axis, H, power.transpose(), vmin = min, vmax = max, zorder=0)
plt.plot( cloudtime, cloudheights, c='r', linewidth=lw, label="cloud tops", zorder=1)
plt.plot( cloudtime, newcloudh, c='c', linewidth=lw, label="Filtered clouds", zorder=2)

plt.scatter(peakx, peakh, c='w', alpha=1, s=100, marker='*', zorder=5, label='peaks')

plt.axhline(y=meanh + meanh/3, c='w', label='peak seperator')
plt.axhline(y=meanh, c='b', label='mean h')

cbar = plt.colorbar(label="Power Ch. 1 (dBz)", mappable= p, ticks=[-10, -15, -20, -25, -30])
# Set the number of ticks
# cbar.locator = ticker.MaxNLocator(nbins=5)
# cbar.update_ticks()

plt.legend(loc='upper right', fontsize=fs/2)
plt.ylabel("Height (m)", fontsize=fs)
plt.xlabel( "Time (Hours, UTC)", fontsize=fs)
ax=plt.gca()
ax.set_facecolor('k')
plt.yticks([1000, 2000, 3000, 4000], fontsize=fs)
plt.xticks(fontsize=fs)
```

```python
# check out the spectra and how effective the response function / filtering is!

window_length = 50 # 150 # 100
T2 = window_length / 2
freq_w = np.arange(0., T2 + 1.) / window_length
P_welch_filt = signal.welch( newcloudh,window='hamming',nperseg=window_length)
P_welch = signal.welch( cloudheights,window='hamming',nperseg=window_length)

P_welch_n_filt = P_welch_filt[1]/np.sum(P_welch_filt[1])  #normalize
### Calculate the response function of the boxcar filter using the convolution theorem
R2_bw=P_welch_filt[1]/P_welch[1]

plt.figure(figsize=(10, 8))
plt.subplot(211)
plt.ylabel('Total Power')
#plt.ylim(0.0,0.40)
plt.xlim(0.0,0.30) 

plt.grid()
plt.plot(freq_w, P_welch[1], 'k', lw=1.5, label = 'Original Spectra') # from above!
plt.plot(freq_w, P_welch_filt[1], 'b', lw=1.5, label = 'Filtered Spectra')
plt.legend(loc='upper right')
plt.title("Total Power Spectrum: WOSA, window length = "+str(window_length),fontsize=14)

plt.subplot(212)
plt.plot(freq_w,R2_bw,'-b',linewidth = 2, label = 'Butterworth N = '+str(N))

plt.xlabel('Frequency (years$^{-1}$)')
plt.ylabel('|R^2| response',fontsize=14)
plt.legend(fontsize = 14, frameon = False)
plt.ylim(0.0,1.2)
plt.grid('on')
plt.xlim(0.0,0.30) 
```
