```python
import os
import pandas as pd
import numpy as np
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/cloud-top-height-stats")
import distributions_general
import distributions_general_flight_legs as xypaths
```

```python
# load all eye passes saved in local dataset
passes = pd.read_pickle("/Users/etmu9498/research/data/all_metadata.pkl")
# passes
```

```python
# composite for two intensity groups (split into strong and weak categories)

# sort for different TC intensities
cutoff = 64 # intensity in kt
tc = 'all' # choose either 'all', '2021', or '2022'
for i in range(2):    
    if i == 0:
        passestrim = passes.drop(index=np.where( passes['intensity'] > cutoff)[0])
        passestrim = passestrim.reset_index()
        title = "Shear Relative Eye Flight Legs: <" + str(cutoff) + ", " + tc
    elif i == 1:
        passestrim = passes.drop(index=np.where( passes['intensity'] <= cutoff)[0])
        passestrim = passestrim.reset_index()
        title = "Shear Relative Eye Flight Legs: >=" + str(cutoff) + ", " + tc
    if not tc.lower() == 'all':
        passestrim = passestrim.drop(index=np.where( ~ ( passestrim['flight'].str[0:4] == tc))[0])
        passestrim = passestrim.reset_index()
    
    print(title)
    print("number of cases: " + str(len(passestrim['flight'].values)))
    res = 60
    xypaths.xyplots( passestrim, res, res, title)
```

```python
# repeat statistic tests for low vs high intensities

for i in range(2):
        
    # too low case
    if i == 0:
        print("Finding Lower Intensity Distributions")
        
        # passes to remove from analysis
        tc = np.where( passes['intensity'] > cutoff )[0].tolist()
        titleend = '< ' + str(cutoff) + " kt"
    # high intensity case
    elif i == 1:
        print("Finding Higher Intensity Distributions")
        
        tc = np.where( passes['intensity'] <= cutoff )[0].tolist()
        titleend = '>= ' + str(cutoff) + " kt"
    
    # different composite tests are listed below
    
    title = "Intensification: " + titleend
    distributions_general.plot_composites(tc=tc, df_passes=passes, bintype='intensification', plottype='averaged', binwidth=25, smoothwidth=25, xaxistype='probability-norm', figtitle=title)
    
    title = "Shear Magnitude: " + titleend
    newlevels=[4.5, 11] # use rogers 2021 definition of shear magnitude
    distributions_general.plot_composites(tc=tc, df_passes=passes, newlevels=newlevels, bintype='shearmag', plottype='averaged', binwidth=25, smoothwidth=25, xaxistype='probability-norm', figtitle=title)    
    
    title = "Shear Quadrant: " + titleend
    distributions_general.plot_composites(tc=tc, df_passes=passes, bintype='sheardir', plottype='averaged', binwidth=25, smoothwidth=25, xaxistype='probability-norm', figtitle=title)

```

```python
# repeat for defined or not eyewalls!

tc = 'all'
for i in range(2):
    if i == 0:
        passestrim = passes.drop(index=np.where( ~ passes['Defined Eyewalls'])[0])
        passestrim = passestrim.reset_index()
        title = "Shear Relative Eye Flight Legs: Defined Eyewalls, " + tc
    elif i == 1:
        passestrim = passes.drop(index=np.where( passes['Defined Eyewalls'])[0])
        passestrim = passestrim.reset_index()
        title = "Shear Relative Eye Flight Legs: Undefined Eyewalls, " + tc
        
    if not tc.lower() == 'all':
        passestrim = passestrim.drop(index=np.where( ~ ( passestrim['flight'].str[0:4] == tc))[0])
        passestrim = passestrim.reset_index()
    
    print(title)
    print("number of cases: " + str(len(passestrim['flight'].values)))
    res = 60
    xypaths.xyplots( passestrim, res, res, title)
```
