```python
# metadata used in both histogram plots
subplot_list = [ 231, 232, 233, 234, 235, 236]
ylabels = [ "Total Wind Speed (m/s)", "Temperature (C)", "WVMR (g/kg)", "Vertical Velocity (m/s)", "Theta E (K)",
          "SFMR Rain Rate (mm/hr)"]
colors = ['b', 'Grey', 'r', 'g', 'y', 'c']
    
binsize = [ 2., .5, .25, .5, 2., 2]
```

```python
###############
## new code
## plot histograms as a colormap to better visualize data!
# this shows many rmws all on one plot, instead of histograms for just one rmw
###############

# using the binsize provided for each case, bin the data manually

# a new data list that includes all bins!
data_list = [ df_binned['wind_speed_array'], df_binned['temp_array'],
        df_binned['Mixing Ratio_array'], df_binned['w_array'], 
        df_binned['Theta E_array'], df_binned['Rain Rate_array'] ]

# a list of colormaps that matches the color of the mean plots! Not all the colors are availible in base matplotlib, though
cmaps = ['Blues', 'Greys', 'Reds', 'Greens', 'YlOrRd', 'Purples']

# add upper limits for each case, and space the contour levels nicely for each case
cbar_lims = [ 110, 50, 30, 100, 80, 90]
levels = ( np.arange(0, cbar_lims[0], step=10), np.arange(0, cbar_lims[1], step=5), np.arange(0, cbar_lims[2], step=5),
         np.arange(0, cbar_lims[3], step=15), np.arange(0, cbar_lims[4], step=10), np.arange(0, cbar_lims[5], step=5),)
# old code: np.append( np.arange(0, cbar_lims[0], step=10), 200.0)

plt.figure( figsize=(18, 12))
helper_fns.change_font_sizes(11, 11)

nancount = 0

# do this for each of the 6 variables
for i in range( len( subplot_list)):

    # find the lowest and highest values for all the bins... use them to make the bin list used below
    tcfield = data_list[i]
    tcfield_list = []
    for fieldi, fieldval in enumerate( tcfield):
        tcfield_list += fieldval
    minval = np.nanmin( tcfield_list)
    maxval = np.nanmax( tcfield_list) 
    
     
    
    # define the x and y axes
    rmws = df_binned[ 'midpoints']
    bins = np.arange( math.floor( minval) - 2*binsize[i], math.ceil( maxval)+ 2*binsize[i], step=binsize[i])
    
    # test things out
    #print( ylabels[ i])
    #print( minval)
    #print(maxval)
    #print( rmws)
    #print( bins)
    #print( len( rmws))
    #print( len( bins))
    #print( np.shape( count_density))
    
    # save the data count here!
    count_density = np.zeros( [ len( rmws), len( bins)])
    
    # print( type( tcfield))
    
    # do this for every rmw bin
    for rmw_i, rmwval in enumerate( rmws):
        # get the local array for this bin
        tcfield_current = tcfield[ rmw_i]
        
        # do this for every value in the locally defined tc field!
        for tcfieldi, tcfieldval in enumerate( tcfield_current):
            
            # find the closest bin for this value
            # negative values after subtraction means we've gone too low: take the lowest positive value
            minusarray = bins - tcfieldval 
          
            # make sure to not index by an empty list!
            if np.all( np.isnan( minusarray)):
                nancount += 1
            else:            
                # find the correct bin index, bin_i, for the closest bin
                # the first [0] looks at the array, and the second [0] takes the first value.
                # The first value is always the smallest!                
                try:
                    bin_i = np.where( minusarray >= 0.0)[0][0]
                    # add one to the nearest bin in the total array!
                    count_density [ rmw_i] [bin_i] += 1
                except:
                    print( minusarray)
            
            # old code, doesn't work :/
            #positives = minusarray[ np.where( minusarray >= 0.0)]
            #print( positives [0])
            #print( np.where( minusarray >= 0.0)[0][0])
            #
            #if len( positives) == 0:
            #    #print( bins - tcfieldval)
            #    #print( positives)
            #    nancount += 1
            #else:            
            #    bin_i = np.argmin( positives)
            #    # add one to the nearest bin in the total array!
            #    count_density [ rmw_i] [bin_i] += 1
                
    
    # make the plot!
    plt.subplot( subplot_list[ i])
    color_map = plt.cm.get_cmap( "RdYlBu").reversed() # plt.cm.get_cmap( cmaps[i]).reversed()

    #plt.pcolormesh( rmws, bins, count_density.transpose(), cmap = color_map, vmin = 0, vmax = cbar_lims[ i]) # vmin = minval - 2 * binsize[ i], vmax = maxval + 2*binsize[ i])
    plt.contourf( rmws, bins, count_density.transpose(), levels[i], cmap = color_map, extend='both')  
    plt.colorbar(label = 'Bin Count') # , spacing='proportional')
    #plt.title( ylabels[ i])
    
    plt.ylabel( ylabels[ i])
    
    #if subplot_list[ i] == 231:
    #    plt.ylabel( "Bin Value")
    if subplot_list[ i] == 235:
        plt.xlabel( "RMW Value")
        # plt.ylim( 340, 380)
    
    
    print( "Subplot " + str( i) + " created")

print( nancount)
#np.set_printoptions(threshold=np.inf)
#print( count_density)
#np.set_printoptions(threshold=1000)
```
