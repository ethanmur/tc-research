```python
# testing the data that brian tang is using!
crlpath = "/Users/etmu9498/research/data/"
data_list = make_plots.load_crl( crlpath, print_files=True)
os.chdir( crlpath)
data = xr.open_dataset( data_list[ 2])

crl_data = data
time = data.time
height = -1000 * data.H
```

```python
plt.figure( figsize=(16, 12))

# plot temperature
plt.subplot( 311)
min = 5
max = 35
map = plt.cm.get_cmap( "RdYlBu").reversed()
plt.pcolormesh( time, height, crl_data.T.transpose(), vmin = min, vmax = max, cmap = map)
plt.colorbar(label="T (Degrees C)")
plt.ylabel("Height (m)")
ax = plt.gca()
ax.set_facecolor('k')

# plot wv
plt.subplot( 312)
min = 0
max = 25
plt.pcolormesh( time, height, crl_data.WVMR.transpose(), vmin = min, vmax = max)
plt.colorbar(label="WVMR (g/kg)")
plt.ylabel("Height (m)")
ax = plt.gca()
ax.set_facecolor('k')

# plot power ch1

plt.subplot( 313)
min = -30
max = -10
plt.pcolormesh( time, height, crl_data.P_ch1.transpose(), vmin = min, vmax = max)
plt.colorbar(label="Power Ch. 1 (dBz)")
plt.ylabel("Height (m)")
ax = plt.gca()
ax.set_facecolor('k')
plt.xlabel( "Time (Hours, UTC)")
```
