```python
# get tc ri names into proper dictionary format
# first, correct the format and put them in a list
intense_list = df['Name'][ np.where( df['Intensification (kt)'].values > 30)[0]].tolist()
for i, val in enumerate( intense_list):
    intense_list[ i] = intense_list[ i][ : -4] + "_processed.nc" # ".nc"

# now, add them to the dict based on year!
# get a list of the years
yearlist = []
filelist = []
for i, val in enumerate( intense_list):
    if len( yearlist) == 0:
        yearlist.append( val[ 0:4])
        filelist.append( [])
    elif not val[0:4] in yearlist:
        yearlist.append( val[ 0:4])
        filelist.append( [])
    else:
        filelist[ -1].append( val )
        
# print( filelist)        
# print( yearlist)

inputdict = {}
# make the dictionary!!
for i, year in enumerate( yearlist):
    inputdict[ year] = filelist[ i]
    
print( inputdict)
```
