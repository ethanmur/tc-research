```python
##########
## 2/23/23 new code
## download track, set, and fix files for all tcs currently present in file structure!!
##########
import os
import urllib
os.chdir( "/Users/etmu9498/research/code/scripts")
import make_plots


# helpful paths
fl_path_root = "/Users/etmu9498/research/data/in-situ-noaa-full/"
data_path = "/Users/etmu9498/research/data/"

# step 1: make a list of year folders and fl files saved in 'in-situ-noaa-full'
os.chdir( data_path)
folder_list = [name for name in os.listdir('in-situ-noaa-full')
    if os.path.isdir(os.path.join('in-situ-noaa-full', name))]


##########
# manually choose a folder list here if you just want to download tracks from certain years!
# make sure that all data is downloaded, though: that's the only way to get the proper names!
##########
# folder_list = ['2012']

total_files = []
# get the unique tc names stored within each year folder
for yeari, yearval in enumerate( folder_list):
    total_files.append( [])
    total_files[yeari] = make_plots.load_flight_level( fl_path_root + yearval, print_files=False) 
    
# step 2: 
# find all the unique names within each year / file folder
unique_names = []
for yeari, yearval in enumerate( folder_list):
    unique_names.append( [])
    for filei, fileval in enumerate( total_files[ yeari]):
        
        # find the underscore and periods, then get the name between these two values!
        # this seems to work well! If it ever breaks down, the names can also be easily isolated 
        # using another method: tcname = fileval[ 11 : -3]
        underscorei = fileval.index( "_")  +1
        doti = fileval.index( ".")
        nameval = fileval[ underscorei : doti]
    
        # empty list case: auto add the name!
        if len( unique_names[ yeari]) == 0:
            unique_names[ yeari].append( nameval)
        # not empty: check if the name exists in the list yet:
        else:
            count = unique_names[ yeari].count( nameval)
            if count == 0:
                unique_names[ yeari].append( nameval)


                
# step 3:
# download the track, set, and fix files!! store them all in the same folder, there shouldn't be too many rn
# maybe change this step later once there are duplicate names!!
trackpath = "/Users/etmu9498/research/data/track"
fixespath = "/Users/etmu9498/research/data/track-fixes"
setspath = "/Users/etmu9498/research/data/track-sets"
for yeari, yearval in enumerate( folder_list):
    print( "Year = " + str( yearval))
    
    for filei, fileval in enumerate( unique_names[ yeari]):

        # case 1: track is before 2018: need to download data from a general folder
        if int( yearval) < 2018:
            dataurl = 'https://www.aoml.noaa.gov/ftp/hrd/data/flightlevel/tracks/'

            # download the correct datasets!
            os.chdir( trackpath)
            try:
                nameval = fileval + yearval + ".trak"
                
                #print( dataurl)
                #print( nameval)
                urllib.request.urlretrieve( dataurl + nameval, nameval)
                print( "track file " + nameval + " downloaded")
            except:
                print( "Unable to download track data for TC " + fileval + ": not sure why")

            os.chdir( fixespath)
            try:
                nameval = fileval + yearval + ".fixs"
                urllib.request.urlretrieve( dataurl + nameval, nameval)
                print( "fix file " + nameval + " downloaded")
            except:
                print( "Unable to download fix data for TC " + fileval + ": not sure why")

                
        elif yearval == '2018' or yearval == '2019':
            print( "Unable to download track data for TC " + fileval + ": not availible for year " + yearval)
            
        elif int( yearval) > 2019:   
            dataurl = 'https://www.aoml.noaa.gov/ftp/hrd/data/flightlevel/' + yearval + '/' + fileval + '/'
   
            # download the correct datasets!
            os.chdir( trackpath)
            try:                
                nameval = fileval + ".trak"
                savename = fileval + yearval + ".trak"
                urllib.request.urlretrieve( dataurl + nameval, fileval + yearval + ".trak")
                print( "track file " + fileval + yearval + " downloaded")
            except:
                print( "Unable to download track data for TC " + fileval + ": not sure why")

            os.chdir( fixespath)
            try:
                nameval = fileval + ".fixs"
                savename = fileval + yearval + ".fixs"
                urllib.request.urlretrieve( dataurl + nameval, fileval + yearval + ".fixs")
                print( "fix file " + fileval + yearval + " downloaded")
            except:
                print( "Unable to download fix data for TC " + fileval + ": not sure why")

        
        
        
```
