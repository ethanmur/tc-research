```python
# make a simple plot to see if time starts to decrease in both arrays (would explain the data overlap)

# make x axes
count7 = np.arange(0, len(crl_1007.time))
count8 = np.arange(0, len(crl_1008.time))
    
print(len(count7))
print(len(count8))
    
plt.figure(figsize=(6,6))
plt.plot(count7, crl_1007.time, c='c', label='10/07')
plt.plot(count8, crl_1008.time, c='r', label='10/08')
plt.legend()
plt.xlabel("Time count since start of recording")
plt.ylabel("Time (UTC, h)")

# plt.xlim([800, 900])
```

```python
# result: one weird spike in time around point 845 is causing this error!
# this makes sense when looking at the crl plots: there's a uniform color overlay stretching to the right side of the
# image, caused by this one outlier time point
```

```python
# fixing the time axis!

time7 = crl_1007.time.values[np.where(crl_1007.time < 28.0)[0]]
time8 = crl_1008.time.values[np.where(crl_1008.time < 28.0)[0]]

# should be one less than array lengths above! looks good
print(len(time7))
print(len(time8))

# actually, instead of deleting the outlier, we want to put a nan there for interpolation. do that below
time7 = np.copy( crl_1007.time.values)
time8 = np.copy( crl_1008.time.values)
time7[np.where(time7 > 28.)[0]] = np.nan
time8[np.where(time8 > 28.)[0]] = np.nan
print(len(time7))
print(len(time8))
# looks perfect!
print(len(np.where(np.isnan(time7))[0]))
print(len(np.where(np.isnan(time8))[0]))
```

```python
# next step: interpolate the one nan value to a time value (decimal)
nans, x = np.isnan(time7), lambda z: z.nonzero()[0]
time7[nans] = np.interp(x(nans), x(~nans), time7[~nans])

nans, x = np.isnan(time8), lambda z: z.nonzero()[0]
time8[nans] = np.interp(x(nans), x(~nans), time8[~nans])
```

```python
# finally, recreate the plot above to show the corrected data!

plt.figure(figsize=(6,6))
plt.plot(count7, time7, c='c', label='10/07')
plt.plot(count8, time8, c='r', label='10/08')
plt.legend()
plt.xlabel("Time count since start of recording")
plt.ylabel("Time (UTC, h)")

# plt.xlim([800, 900])
```

```python
# make pretty time series plots to confirm that this worked!

timelist = [time7, time8]
crl_list = [ crl_1007, crl_1008]
for timei, time in enumerate(timelist):
    crl_data = crl_list[timei]

    fig = plt.figure( figsize=(18, 12), facecolor='w')
    fs = 15
    helper_fns_winter2023.change_font_sizes(fs, fs)

    # plot temperature with original distance axis
    plt.subplot(411)
    # plot power
    power = 10 * np.log10( crl_data.P_ch1.transpose() )
    power = power.where( power.values > -40)
    power = power.where( power.values < 0)

    plt.pcolormesh(  time, -crl_data.H, power, vmin = -40, vmax =-10)    
    plt.ylabel( 'Height (km)')    
    cbar = plt.colorbar()
    cbar.ax.set_ylabel( ylabel="CRL Return Power (dBz)")
    cbar.ax.tick_params(labelsize=fs)
    plt.grid( 'on')
    ax = plt.gca()
    ax.set_facecolor('k')

    # plot T
    plt.subplot(412)
    color_map = plt.cm.get_cmap( "RdYlBu").reversed()
    plt.pcolormesh( time, -crl_data.H, crl_data.T.transpose(), cmap = color_map, vmin=5, vmax=25 )
    cbar = plt.colorbar()
    cbar.ax.set_ylabel( ylabel="CRL T ( C)")
    cbar.ax.tick_params(labelsize=fs)
    plt.ylabel( 'Height (km)')
    plt.grid( 'on')
    ax = plt.gca()
    ax.set_facecolor('k')

    # plot wvmr
    plt.subplot(413)
    plt.pcolormesh( time, -crl_data.H, crl_data.WVMR.transpose(), vmin = 8, vmax =18)
    cbar = plt.colorbar()
    cbar.ax.set_ylabel( ylabel="CRL WVMR (g/Kg)")
    cbar.ax.tick_params(labelsize=fs)
    plt.ylabel( 'Height (km)')
    plt.grid( 'on')
    ax = plt.gca()
    ax.set_facecolor('k')

    # plot light scattering ratio
    plt.subplot(414)
    plt.pcolormesh( time, -crl_data.H, crl_data.LSR.transpose(), vmin = 0, vmax =10)
    cbar = plt.colorbar()
    cbar.ax.set_ylabel( ylabel="CRL WVMR (g/Kg)")
    cbar.ax.tick_params(labelsize=fs)
    plt.ylabel( 'Height (km)')
    plt.grid( 'on')
    ax = plt.gca()
    ax.set_facecolor('k')

    plt.xlabel( "Time (UTC, Hours)")
    plt.show()
    
    print("plot " + str(timei) + ' created')
```
