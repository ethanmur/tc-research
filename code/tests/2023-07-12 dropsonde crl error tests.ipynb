```python
# use the plot helper function to find standard deviations at evenly spaced height bins

currentbin = 0.
binmax = 3100.
binwidth = 100.

binlist = []
stdlist_t = []
stdlist_wv = []
count_t = []
count_wv = []
while True:
    if currentbin+binwidth > binmax:
        break
    hlims = [currentbin, currentbin+binwidth]    
    allcrlt, allcrlwv, allsondet, allsondewv, tanom, wvanom, theights, wvheights = sonde_crl_errors.plot_helper(df, height_limit=hlims)

    # testing
    # print(np.nanmax(np.array(theights)))
    # print(np.nanmin(np.array(theights)))
    
    # save standard deviations and data count for each bin
    stdlist_t.append( np.std(tanom))
    stdlist_wv.append( np.std(wvanom))
    count_t.append(len(tanom))
    count_wv.append(len(wvanom))
    binlist.append(currentbin)
    
    # iterate the bins for next run 
    currentbin = currentbin + binwidth

    
plt.figure(figsize=(11, 8))

plt.subplot(411)
plt.plot(binlist, count_t, c='r')
plt.ylabel("T Comparison Count")
plt.gca().invert_xaxis()

plt.subplot(412)
plt.plot(binlist, count_wv, c='g')
plt.ylabel("WV Comparison Count")
plt.gca().invert_xaxis()

plt.subplot(413)
plt.plot(binlist, stdlist_t, c='r')
plt.ylabel("St Dev T (C)")
plt.gca().invert_xaxis()

plt.subplot(414)
plt.plot(binlist, stdlist_wv, c='r')
plt.ylabel("St Dev WV (g/kg)")
plt.xlabel("Bin Start Value (m)")
plt.gca().invert_xaxis()
```

```python
# loop through all dropsondes and count the number of non-nan points

totalcount = 0 # all possible data points (including nans)
crlcount = 0 # all non nan crl data points
sondecount = 0 # all non nan sonde data points
nonans = 0 # all heights with non nan crl and sonde points

for rowi in range(len(df['year'])):
    totalcount += len(df['height (m)'][rowi])
    
    crlt = df['crl t (C)'][rowi]
    sondet = df['sonde t (C)'][rowi]
    
    crlcount += len( np.where( ~np.isnan(crlt))[0])
    sondecount += len( np.where( ~np.isnan(sondet))[0])
    
    nonans += len( np.intersect1d( np.where( ~np.isnan(crlt))[0], np.where( ~np.isnan(sondet))[0]))
    
print("Number of possible crl sonde combos: " + str( totalcount))
print("Number of valid crl data points:     " + str( crlcount))
print("Number of valid sonde data points:   " + str(sondecount))
print("Number of valid crl sonde combos:    " + str(nonans))

print("Number of only sonde data points impacting nonan:     " + str(crlcount-nonans))
print("Number of sonde nans:     " + str(totalcount-sondecount))
```
