```python
import os
import numpy as np
os.chdir("/Users/etmu9498/research/code/scripts-winter2023/cloud-top-height-stats/")
import cloud_top_plotting_individual as plot

inputdict = {}
inputdict['2021'] = ['P3_20210821H2_processed.nc', 'P3_20210929H2_processed.nc', 'P3_20210820H1_processed.nc']

df_peaks = plot.plot_one_day(tc='all')
```

```python
print('total dates in dataset: {}'.format(len( df_peaks['year'])))

count2021 = len(df_peaks['year'][np.where(df_peaks['year']=='2021')[0]])
count2022 = len(df_peaks['year'][np.where(df_peaks['year']=='2022')[0]])
print('dates in 2021: {}'.format(count2021))
print('dates in 2022: {}'.format(count2022))

counti = list(np.arange(0,6))
weak_count = []
strong_count = []
for count_i in range( 6):
    peak_count_i = len(df_peaks['peak counts strong'][np.where(df_peaks['peak counts strong'] == count_i)[0]])
    strong_count.append(peak_count_i)
    print('number of cases with {} strong peaks: {}'.format(count_i, peak_count_i) )    

    peak_count_i = len(df_peaks['peak counts weak'][np.where(df_peaks['peak counts weak'] == count_i)[0]])
    weak_count.append(peak_count_i)
    print('number of cases with {} weak peaks: {}'.format(count_i, peak_count_i) )    

    
print("Note: cases with 0 peaks signify no successful eye passes for that day!")
    
print(counti)
print(strong_count)
print(weak_count)

df_peaks
```

```python
# find the mean peak height when only one peak is present
heights = np.array( df_peaks['peak heights (m)'][np.where(df_peaks['peak counts strong'] == 1)[0]])
# flatten the list of lists!
heights = [item for sublist in heights for item in sublist]

print( heights)

print( 'mean heights for 1 strong peak: {}'.format(np.mean(heights)))

# repeat this for lower height peaks for 2 to 3 peak cases
heights = np.array( df_peaks['peak heights (m)'][np.where( np.logical_or(df_peaks['peak counts strong'] == 2, df_peaks['peak counts strong'] == 3))[0]])
print(heights)
# flatten the list of lists! only keep the first aka lowest height
tempheights = []
for sublist in heights:
    tempheights.append(sublist[0])
heights = tempheights

print( 'mean lower heights: {}'.format(np.mean(heights)))

# repeat this for higher height peaks for 2 to 3 peak cases
heights = np.array( df_peaks['peak heights (m)'][np.where( np.logical_or(df_peaks['peak counts strong'] == 2, df_peaks['peak counts strong'] == 3))[0]])
print(heights)
# flatten the list of lists! only keep the first aka lowest height
tempheights = []
for sublist in heights:
    tempheights.append(sublist[-1])
heights = tempheights

print( 'mean higher heights: {}'.format(np.mean(heights)))
```
